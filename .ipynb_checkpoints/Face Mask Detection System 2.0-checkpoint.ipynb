# Face Mask Detection System

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import cv2
import tensorflow as tf
import tensorflow.keras as keras
```

### Load Data set

```python
dataPath='dataset' #Dataset which has two folders(with_mask,without_mask)
classFile = os.listdir(dataPath)
labels=[i for i in range(len(classFile))]
label_dict=dict(zip(classFile,labels))
```

### Process the data

```python
data=[]
target_label=[]

for file in classFile:
    folderPath=os.path.join(dataPath, file)
    png_names= os.listdir(folderPath)
    
    for png_name in png_names:
        image_path=os.path.join(folderPath,png_name)
        img=cv2.imread(image_path)
        
        try:
            grayed_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            resized_img=cv2.resize(grayed_img,(100,100))
            data.append(resized_img)
            target_label.append(label_dict[file])
        
        except Exception as e:
            print('Exception:',e)
```

```python
from keras.utils import np_utils

data=np.array(data)/255.0
data=np.reshape(data,(data.shape[0],100,100,1))
target_label=np.array(target_label)

newTarget = np_utils.to_categorical(target_label)

np.save('data',data)
np.save('target',newTarget)
```

### Build algorithms

```python
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras import optimizers

algo = models.Sequential()
algo.add(layers.Conv2D(32, (3, 3), activation='relu',input_shape=data.shape[1:]))
algo.add(layers.MaxPooling2D((2, 2)))
algo.add(layers.Conv2D(64, (3, 3), activation='relu'))
algo.add(layers.MaxPooling2D((2, 2)))
algo.add(layers.Conv2D(128, (3, 3), activation='relu'))
algo.add(layers.MaxPooling2D((2, 2)))
algo.add(layers.Conv2D(128, (3, 3), activation='relu'))
algo.add(layers.MaxPooling2D((2, 2)))
algo.add(layers.Flatten())
algo.add(layers.Dropout(0.5))
algo.add(layers.Dense(512, activation='relu'))
algo.add(layers.Dense(1, activation='sigmoid'))
```

```python
algo.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

```python
from sklearn.model_selection import train_test_split
trainData,testData,trainTarget,testTarget = train_test_split(data,target_label,test_size=0.1)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

checkpoint = ModelCheckpoint(
  'model-{epoch:03d}.model',
  monitor='val_loss',
  verbose=0,
  save_best_only=True,
  mode='auto')

history=algo.fit(
  trainData,
  trainTarget,
  epochs=15,
  callbacks=[checkpoint],
  validation_split=0.2)
```

### Analysis Data

```python
pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1)
plt.show()

figure = plt.gcf()
figure.set_size_inches(8,6)
```

```python
print(algo.evaluate(testData,testTarget))
```

##### Accuracy = 94.20%
